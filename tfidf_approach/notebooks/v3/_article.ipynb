```python
import re
import html

def clean_html_basic(html_text: str) -> str:
    if not isinstance(html_text, str):
        return html_text

    # 1) rimuovi script e style
    html_text = re.sub(
        r"<(script|style).*?>.*?</\1>",
        " ",
        html_text,
        flags=re.DOTALL | re.IGNORECASE
    )

    # 2) rimuovi tutti i tag HTML
    html_text = re.sub(r"<[^>]+>", " ", html_text)

    # 3) decodifica entit√† HTML (&nbsp;, &amp;, ...)
    html_text = html.unescape(html_text)

    # 4) normalizza spazi
    html_text = re.sub(r"\s+", " ", html_text)

    return html_text.strip()
```

```python
df['article'] = df['article'].apply(clean_html_basic)
```
