```python
from src.utils import load_data

df = load_data(DEVELOPMENT_PATH)
```

```python
df['y'].value_counts() / df.shape[0] * 100
```

```python
import matplotlib.pyplot as plt

counts = df["y"].value_counts().sort_index()
perc = (counts / counts.sum() * 100)

label_map = {
    0: "International",
    1: "Business",
    2: "Technology",
    3: "Entertainment",
    4: "Sports",
    5: "General",
    6: "Health",
}

x = [label_map.get(i, str(i)) for i in counts.index]

plt.figure()
plt.bar(x, counts.values)
plt.ylabel("Count")
plt.xlabel("Class")
plt.xticks(rotation=30, ha="right")
plt.grid(axis="y", linestyle="--", linewidth=0.5, alpha=0.6)
plt.tight_layout()
plt.savefig("target_distribution_counts.pdf", bbox_inches="tight")
plt.show()

plt.figure()
plt.bar(x, perc.values)
plt.ylabel("Percentage (%)")
plt.xlabel("Class")
plt.xticks(rotation=30, ha="right")
plt.grid(axis="y", linestyle="--", linewidth=0.5, alpha=0.6)
plt.tight_layout()
plt.savefig("target_distribution_percent.pdf", bbox_inches="tight")
plt.show()

summary = pd.DataFrame({"count": counts, "percent": perc.round(2)})
print(summary)
```

It looks like an unbalanced problem, so this is something that we should take into account in the next phase.
