```python
from src.utils import load_data
from src.config import DEVELOPMENT_PATH
news_df = load_data(DEVELOPMENT_PATH)
```

```python
from collections import Counter
from nltk.tokenize import word_tokenize

word_freq = Counter()

sub_df = news_df[news_df['y'] == 0]

sub_df["article"].apply(
    lambda x: word_freq.update(
        t for t in word_tokenize(x.lower())
        if t.isalpha() and len(t) > 2
    )
);
top0_50 = word_freq.most_common(50);

```

```python
word_freq = Counter()

sub_df = news_df[news_df['y'] == 3]

sub_df["article"].apply(
    lambda x: word_freq.update(
        t for t in word_tokenize(x.lower())
        if t.isalpha() and len(t) > 2
    )
);
top3_50 = word_freq.most_common(50)
```

```python
word_freq = Counter()

sub_df = news_df[news_df['y'] == 5]

sub_df["article"].apply(
    lambda x: word_freq.update(
        t for t in word_tokenize(x.lower())
        if t.isalpha() and len(t) > 2
    )
);
top5_50 = word_freq.most_common(50)
```

```python
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
```

```python
missing_sw_0 = [word[0] for word in top0_50 if word[0] not in ENGLISH_STOP_WORDS]
missing_sw_3 = [word[0] for word in top3_50 if word[0] not in ENGLISH_STOP_WORDS]
missing_sw_5 = [word[0] for word in top5_50 if word[0] not in ENGLISH_STOP_WORDS]

common_sw = [word for word in missing_sw_0 if word in missing_sw_3]
common_sw = [word for word in common_sw if word in missing_sw_5]
```
