```python
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay

# Normalize row-wise
cm_old_norm = cm_old / cm_old.sum(axis=1, keepdims=True)
cm_final_norm = cm_final / cm_final.sum(axis=1, keepdims=True)

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

disp_old = ConfusionMatrixDisplay(
    confusion_matrix=cm_old_norm,
    display_labels=range(0,7)
)
disp_old.plot(ax=axes[0], values_format=".2f", colorbar=False)
axes[0].set_title("Before class weighting")

disp_final = ConfusionMatrixDisplay(
    confusion_matrix=cm_final_norm,
    display_labels=range(0,7)
)
disp_final.plot(ax=axes[1], values_format=".2f", colorbar=False)
axes[1].set_title("After class weighting")

plt.tight_layout()
plt.savefig("confusion_matrices_comparison.pdf", bbox_inches="tight")
plt.show()
```
