Class weights experiments summary.

1) class_weight="balanced":
Using inverse-frequency class weights provides a solid and principled baseline for macro-F1 optimization in the presence of class imbalance.
Results are stable and improve performance on the rarest class, but overall gains are moderate.

2) Clipped class weights based on label distribution:
Clipping inverse-frequency weights (e.g. limiting the rarest class weight) represents a good trade-off between recall improvement and stability.
This configuration yields consistent improvements over the balanced baseline and remains well-justified by the empirical label distribution.

3) Class weights directly proportional to inverse class frequencies (no clipping):
This setup achieves the best macro-F1 observed so far (~0.732 on public score), mainly by significantly improving performance on the rarest class.
The improvement is coherent with the macro-F1 objective and does not severely degrade other classes, making this configuration the current best candidate.

4) Aggressively increased weights for the most confused class:
Manually boosting the weight of the most confused class can further improve validation performance in some runs.
However, this approach is highly sensitive to the validation split and shows worse generalization on the public leaderboard (~0.725).
Therefore, it is considered prone to overfitting and not fully defensible, and it is not retained as the final configuration.
